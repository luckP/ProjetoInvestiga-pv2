```python
import pandas as pd
import numpy as np
import datetime as dt
import time
import random


# This import registers the 3D projection, but is otherwise unused.
from mpl_toolkits.mplot3d import Axes3D  # noqa: F401 unused import

import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter

print 'ok'
```

```python
st = time.time()
df = pd.read_csv("../serve/df.csv") 

print time.time() - st
df.head()
```

```python
# variaveis de controle

squares = df.id_square.unique()
minutes = 15
time_window = 60*minutes
square_id = 1
```

```python
# Criada a lista de timewindows
t = time.time()

t_min =  df.events_timestamp.min()
t_max =  df.events_timestamp.max()
model_df = pd.DataFrame({  int( (t_min + time_window*j)/time_window)*time_window   : {'in': 0, 'out': 0, 'out_error': 0} for j in range(int((t_max - t_min)/time_window)+1)})
model_df = model_df.T
print time.time() - t
```

```python
square_list = { i: model_df.copy() for i in range(1, 65) }

print 'ok'
```

```python
df[df['id_taxi']==20007073].head()
```

```python
st = time.time()
remove_in = {}
    
for row in df.itertuples():
#   SUBISTITUIR O RANDOM PELO VALOR CORRETO
    id_square = random.randint(1,64)
    index = int(row.events_timestamp/time_window)*time_window
    
    if row.type == 'stop':
        square_list[id_square]['in'][index]+=1 
        if row.id_taxi in remove_in and remove_in[row.id_taxi]>0:
#             REMOVER IF
            if remove_in[row.id_taxi]>1:
                print 'error, teoricamente nao era para esse valor ser maior que 1'
            remove_in[row.id_taxi]-=1
            square_list[id_square]['in'][index]-=1 
    elif row.type == 'recolha' or row.type == 'busy':
        square_list[id_square]['out'][index]+=1
    elif row.type == 'free':
        square_list[id_square]['out_error'][index]+=1
    else:
        if not row.id_taxi in remove_in:
            remove_in[row.id_taxi] = 1
        else:
            remove_in[row.id_taxi]+=1
        
print time.time() - st
```

```python
# Data for plotting
# y = [data[1][x]['in'] for x in data[1]]
x = []
y_in = []
y_out = []
y_out_error = []

st = time.time()

# for t in square_list[1]:
#     x.append(t)
#     y_in.append(data[square_id][t]['in'])
#     y_out.append(data[square_id][t]['out'])
#     y_out_error.append(data[square_id][t]['out_error'])

x = [int( (t_min + time_window*j)/time_window)*time_window for j in range(int((t_max - t_min)/time_window)+1)]
y_in = [i for i in square_list[1]['in']]
y_out = [i for i in square_list[1]['out']]
y_out_error = [i for i in square_list[1]['out_error']]

print time.time() - st
```

```python

fig, ax = plt.subplots()
fig.set_size_inches(800, 5)

ax.plot( x, y_in, color='#4a69bd', linewidth=1)
ax.plot( x, y_out,  color='#78e08f', linewidth=1)
ax.plot( x, y_out_error, color='#e55039', linewidth=1)

fig.savefig(str(square_id)+".png", dpi = 72) 
# plt.show()
```

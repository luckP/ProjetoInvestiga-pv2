```python
import pandas as pd
import numpy as np
import datetime as dt
import time


# This import registers the 3D projection, but is otherwise unused.
from mpl_toolkits.mplot3d import Axes3D  # noqa: F401 unused import

import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter

print 'ok'
```

```python
st = time.time()
df = pd.read_csv("../serve/df.csv") 

print time.time() - st
df.head()
```

```python
# variaveis de controle

squares = df.id_square.unique()
minutes = 15
time_window = 60*minutes
square_id = 1159
```

```python
st = time.time()

data = {}
remove_in = {}

for square in squares:
    data[square] = {}
    
for row in df.itertuples():
    if not int(row.events_timestamp/time_window) in data[row.id_square]:
        data[row.id_square][int(row.events_timestamp/time_window)] = {'in': 0, 'out': 0, 'out_error': 0}
        
    if row.type == 'stop':
        data[row.id_square][int(row.events_timestamp/time_window)]['in']+=1 
        if row.id_taxi in remove_in and remove_in[row.id_taxi]>0:
#             REMOVER IF
            if remove_in[row.id_taxi]>1:
                print 'error, teoricamente nao era para esse valor ser maior que 1'
            remove_in[row.id_taxi]-=1
            data[row.id_square][int(row.events_timestamp/time_window)]['in']-=1 
    elif row.type == 'recolha' or row.type == 'busy':
        data[row.id_square][int(row.events_timestamp/time_window)]['out']+=1
    elif row.type == 'free':
        data[row.id_square][int(row.events_timestamp/time_window)]['out_error']+=1
    else:
        if not row.id_taxi in remove_in:
            remove_in[row.id_taxi] = 1
        else:
            remove_in[row.id_taxi]+=1
        
print time.time() - st

data[square_id]
```

```python
# Data for plotting
# y = [data[1][x]['in'] for x in data[1]]
x = []
y_in = []
y_out = []
y_out_error = []

st = time.time()

list = [t for t in data[square_id]]

for t in list:
    x.append(t)
    y_in.append(data[square_id][t]['in'])
    y_out.append(data[square_id][t]['out'])
    y_out_error.append(data[square_id][t]['out_error'])

print time.time() - st


len(y_in)
fig, ax = plt.subplots()
fig.set_size_inches(200, 5)

# ax.plot(x, y)

ax.plot(x, y_in, color='#4a69bd', linewidth=1)
ax.plot( x, y_out,  color='#78e08f', linewidth=1)
ax.plot( x, y_out_error, color='#e55039', linewidth=1)

fig.savefig(str(square_id)+".png", dpi = 72) 
# plt.show()
```

```python
fig = plt.figure()
ax = fig.gca(projection='3d')

# Make data.
X = np.arange(0, 24, 0.25)
Y = np.arange(0, 8760)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = []
for i in X:
    for j in Y:
        Z.append(X*Y)

# Plot the surface.
surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)

# Customize the z axis.
ax.set_zlim(-1.01, 1.01)
ax.zaxis.set_major_locator(LinearLocator(10))
ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))

# Add a color bar which maps values to colors.
fig.colorbar(surf, shrink=0.5, aspect=5)

plt.show()
        
```
